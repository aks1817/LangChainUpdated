##### OpenAI Model

### Streaming And Batch

#### Streaming

Most models can stream their output content while it is being generated. By displaying output progressively, streaming significantly improves user experience, particularly for longer responses. Calling stream() returns am iteratir that yields chunks as they are produced. You can use a loop to process each chunk in real-time

```python
for chunk in model.stream("Write me a 200 words paragraph on Artificial Intelligence in Healthcare."):
    print(chunk.content, end='', flush=True)
```

#### Batch

Bacthing a collection of independent requests to a model can significantly improve performance and reduce costs, as the processing ca be done in parallel:

```python
response=model.batch(["Hello, how are you?", "What is the capital of France?"],
config={
    "max_concurrency":5, # Limit to 5 parallel requests
})

for res in response:
    print(res.content)
```
